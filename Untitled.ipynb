```python
import numpy as np
import numpy as np, scipy.stats as st
import collections

def get_intervals(in_data):
    ret = st.norm.interval(0.95, loc=np.mean(in_data), scale=st.sem(in_data))
    return ret

def calc_confidence_intervals(i,in_X,in_y):
    conf_interval_per_node =  collections.defaultdict(list)
    cur_labels = np.unique(in_y)
    for label in cur_labels:
        label_rows = np.where(in_y==label)
        # print('label',label)
        # print('label_rows',label_rows)
        # print(label)
        # print(np.apply_along_axis(get_intervals, 0, clf.samples[label_rows]))
        # print('----')
        conf_interval_per_node[label] = np.apply_along_axis(get_intervals, 0, clf.samples[label_rows])
    # print(conf_interval_per_node)
    return conf_interval_per_node

def calc_confidence_intervals_all():
    confidence_intervals_per_node = collections.defaultdict(list)
    for i in range(clf.tree_.node_count):
        # print(clf.samples[clf.samples_per_node[i]])
        # print(clf.labels[clf.samples_per_node[i]])
        confidence_intervals_per_node[i] = calc_confidence_intervals(i,clf.samples[clf.samples_per_node[i]],clf.labels[clf.samples_per_node[i]])
    return confidence_intervals_per_node
```
